Now we'd like to find out why these books are the ones that were returned when we searched the LTS bibliography for "eutopia":

```python
print("number of eutopia list books:")
l1 = len(eutopia_df)
print(l1)
print("number of eutopia list books with 'eutopia' or 'Eutopia' in the abstract:")
l2 = len(eutopia_df[(eutopia_df.abstract.str.contains('eutopia')) | (eutopia_df.abstract.str.contains('Eutopia'))])
print(l2)
print("number of eutopia list books w/o 'eutopia' or 'Eutopia' in the abstract:")
print(l1-l2, '\n')

print("if 'eutopia' or 'Eutopia' aren't in the abstract, where are they?")
mask = (~((eutopia_df.abstract.str.contains('eutopia')) | (eutopia_df.abstract.str.contains('Eutopia'))))
for column in eutopia_df.columns:
    if column == 'abstract':
        continue
    print(column)
    print(eutopia_df[mask & (eutopia_df[column].astype(str).str.contains('eutopia') | eutopia_df[column].astype(str).str.contains('Eutopia'))][column].values)
```

Ok, so 4 of the works w/o E/eutopia in the abstract have it in the title, but that still leaves 99 works unaccounted for.

```python
def clean_epub(path):
    book = epub.read_epub(path)
    html_corpus = []
    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            html_corpus.append(item.get_content())
    cleaned = html2text(html_corpus) 
    return cleaned
    
    
```

```python
def chap2text(chap):
    output = ''
    blacklist = ['[document]',   'noscript', 'header',  'html', 'meta', 'head','input', 'script']
    soup = BeautifulSoup(chap, 'html.parser')
    text = soup.find_all(text=True)
    for t in text:
        if t.parent.name not in blacklist:
            output += '{} '.format(t)
    return output
```

```python
def html2text(thtml):
    Output = []
    for html in thtml:
        text =  chap2text(html)
        Output.append(text)
    return Output
```

```python
def clean_texts(texts):
    cleaned_texts = []
    cleaned_text = []
    
    for text in texts:
        doc = nlp(text.lower())
        for w in doc:
            # if it's not a stop word or punctuation mark, add it to our article!
            if w.text != '\n' and not w.is_stop and not w.is_punct and not w.like_num and w.text != 'I':
                # we add the lematized version of the word
                cleaned_text.append(str(w).strip())
        cleaned_texts.append(cleaned_text)
        cleaned_text = []
        
    bigram = gensim.models.Phrases(cleaned_texts)
    cleaned_texts = [bigram[line] for line in cleaned_texts]
    cleaned_words = []
    for text in cleaned_texts:
        for word in text:
            cleaned_words.append(word)
    return cleaned_words
```

```python
def chunks(l, n):
    n = max(1, n)
    return list((l[i:i+n] for i in range(0, len(l), n)))
```

## Dystopia Texts

```python
raw_texts_dystopia = {}
```

```python
directory = "Dystopia Corpus Full Texts"
```

```python
for filename in os.listdir(directory):
    if filename not in raw_texts_dystopia:
        try:
            if filename.endswith(".txt"):
                raw_texts_dystopia[filename] = clean_text(directory + "/" + filename)
                continue
            if filename.endswith(".epub"):
                raw_texts_dystopia[filename] = clean_epub(directory + "/" + filename)
                continue
            if filename.endswith(".pdf"):
                raw_texts_dystopia[filename] = clean_pdf(directory + "/" + filename)
        except:
            print("Error with file " + filename)
```
